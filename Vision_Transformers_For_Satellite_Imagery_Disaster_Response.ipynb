```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
import matplotlib.pyplot as plt
import random
import os

# Set random seeds for reproducibility
tf.random.set_seed(42)
np.random.seed(42)
random.seed(42)

# Parameters
IMG_SIZE = 64
BATCH_SIZE = 16
EPOCHS = 10
NUM_CLASSES = 2  # Low risk, High risk

# Map some EuroSAT classes to disaster risk level - for illustration
# EuroSAT classes (RGB):
# ['AnnualCrop', 'Forest', 'HerbaceousVegetation', 'Highway', 'Industrial', 'Pasture', 'PermanentCrop', 'Residential', 'River', 'SeaLake']
# Let's say 'Forest', 'Pasture', 'PermanentCrop', 'River', 'SeaLake' = Low risk
# 'AnnualCrop', 'HerbaceousVegetation', 'Highway', 'Industrial', 'Residential' = High risk

LOW_RISK_CLASSES = ['Forest', 'Pasture', 'PermanentCrop', 'River', 'SeaLake']
HIGH_RISK_CLASSES = ['AnnualCrop', 'HerbaceousVegetation', 'Highway', 'Industrial', 'Residential']

# For safe distances in meters for each risk level:
SAFE_DISTANCE = {
    'Low Risk': 50,   # safe distance in meters
    'High Risk': 200
}

print("Loading EuroSAT dataset from TensorFlow Datasets...")

import tensorflow_datasets as tfds

# Load EuroSAT dataset, RGB images
dataset, info = tfds.load('eurosat/rgb', split='train', with_info=True, as_supervised=True)

# We will filter and limit our dataset to 100-200 samples total, balanced between low and high risk classes.

class_names = info.features['label'].names

# Reverse mapping from class name to label
class_name_to_label = {name: idx for idx, name in enumerate(class_names)}

# We want to only keep classes that are mapped to low or high risk
low_risk_labels = [class_name_to_label[c] for c in LOW_RISK_CLASSES if c in class_name_to_label]
high_risk_labels = [class_name_to_label[c] for c in HIGH_RISK_CLASSES if c in class_name_to_label]

# Separate dataset into low and high risk class subsets
low_risk_ds = dataset.filter(lambda img, label: tf.reduce_any(label == low_risk_labels))
high_risk_ds = dataset.filter(lambda img, label: tf.reduce_any(label == high_risk_labels))

# Function to sample N examples from dataset
def sample_n(ds, n):
    ds = ds.shuffle(1000, seed=42)
    return ds.take(n)

# Take ~100 images from low risk and ~100 from high risk
low_risk_count = 100
high_risk_count = 100
low_risk_sampled = sample_n(low_risk_ds, low_risk_count)
high_risk_sampled = sample_n(high_risk_ds, high_risk_count)

# Merge datasets
full_ds = low_risk_sampled.concatenate(high_risk_sampled)

# Function to preprocess and label disaster risk
def preprocess_img_label(image, label):
    # Resize images to IMG_SIZE x IMG_SIZE
    image = tf.image.resize(image, [IMG_SIZE, IMG_SIZE])
    image = tf.cast(image, tf.float32) / 255.0  # normalize to [0,1]

    # Map the original label to 0 (low risk) or 1 (high risk)
    def map_to_binary(label):
        return tf.cond(tf.reduce_any(label == low_risk_labels),
                       lambda: tf.constant(0, dtype=tf.int64),
                       lambda: tf.constant(1, dtype=tf.int64))

    # Because label is a scalar tensor, we can use tf.py_function:
    label_binary = tf.py_function(lambda x: 0 if x in low_risk_labels else 1, [label], tf.int64)
    label_binary.set_shape([])

    return image, label_binary

# Map preprocessing
full_ds = full_ds.map(preprocess_img_label, num_parallel_calls=tf.data.AUTOTUNE)
full_ds = full_ds.shuffle(200, seed=42)

# Split into train and validation (85% / 15%)
total_count = low_risk_count + high_risk_count
train_count = int(total_count * 0.85)
val_count = total_count - train_count

train_ds = full_ds.take(train_count).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)
val_ds = full_ds.skip(train_count).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)

print(f"Using {total_count} images: {train_count} train, {val_count} val")

# Define Vision Transformer components

class PatchExtract(layers.Layer):
    def __init__(self, patch_size):
        super().__init__()
        self.patch_size = patch_size

    def call(self, images):
        batch_size = tf.shape(images)[0]
        patches = tf.image.extract_patches(
            images=images,
            sizes=[1, self.patch_size, self.patch_size, 1],
            strides=[1, self.patch_size, self.patch_size, 1],
            rates=[1,1,1,1],
            padding='VALID'
        )
        patch_dims = patches.shape[-1]
        patches = tf.reshape(patches, [batch_size, -1, patch_dims])
        return patches

class PatchEncoder(layers.Layer):
    def __init__(self, num_patches, projection_dim):
        super().__init__()
        self.num_patches = num_patches
        self.projection = layers.Dense(units=projection_dim)
        self.position_embedding = layers.Embedding(
            input_dim=num_patches, output_dim=projection_dim
        )

    def call(self, patch):
        positions = tf.range(start=0, limit=self.num_patches, delta=1)
        encoded = self.projection(patch) + self.position_embedding(positions)
        return encoded

def create_vit_classifier(
    input_shape=(IMG_SIZE, IMG_SIZE, 3),
    patch_size=8,
    projection_dim=64,
    num_heads=2,
    transformer_layers=4,
    mlp_head_units=[128, 64],
    num_classes=NUM_CLASSES,
):
    inputs = layers.Input(shape=input_shape)
    # Create patches.
    patches = PatchExtract(patch_size)(inputs)
    num_patches = (input_shape[0] // patch_size) * (input_shape[1] // patch_size)
    # Encode patches.
    encoded_patches = PatchEncoder(num_patches, projection_dim)(patches)

    # Create multiple Transformer blocks.
    for _ in range(transformer_layers):
        # Layer normalization 1.
        x1 = layers.LayerNormalization(epsilon=1e-6)(encoded_patches)
        # Create a multi-head attention layer.
        attention_output = layers.MultiHeadAttention(
            num_heads=num_heads, key_dim=projection_dim, dropout=0.1
        )(x1, x1)
        # Skip connection 1.
        x2 = layers.Add()([attention_output, encoded_patches])
        # Layer normalization 2.
        x3 = layers.LayerNormalization(epsilon=1e-6)(x2)
        # MLP.
        mlp_output = keras.Sequential(
            [
                layers.Dense(units=projection_dim * 2, activation=tf.nn.gelu),
                layers.Dense(units=projection_dim, activation=tf.nn.gelu),
            ]
        )(x3)
        # Skip connection 2.
        encoded_patches = layers.Add()([mlp_output, x2])

    # Create a [batch_size, projection_dim] tensor.
    representation = layers.LayerNormalization(epsilon=1e-6)(encoded_patches)
    representation = layers.Flatten()(representation)
    representation = layers.Dropout(0.5)(representation)

    # MLP head.
    for units in mlp_head_units:
        representation = layers.Dense(units, activation=tf.nn.gelu)(representation)
        representation = layers.Dropout(0.3)(representation)

    logits = layers.Dense(num_classes, activation='softmax')(representation)

    # Create the Keras model.
    model = keras.Model(inputs=inputs, outputs=logits)
    return model

# Create the model
model = create_vit_classifier()
model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=1e-3),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy'],
)

print(model.summary())

# Train the model
print("Starting model training...")
history = model.fit(train_ds, epochs=EPOCHS, validation_data=val_ds)

# Function to map predicted label to risk level and safe distance
def get_risk_and_distance(pred_probs):
    pred_label = np.argmax(pred_probs)
    if pred_label == 0:
        risk = "Low Risk"
    else:
        risk = "High Risk"
    distance = SAFE_DISTANCE[risk]
    return risk, distance

# Evaluate some samples from validation set and display
print("\nSample predictions on validation set:")
val_iter = iter(val_ds.unbatch().batch(5))
images_batch, labels_batch = next(val_iter)
predictions = model.predict(images_batch)
for i in range(len(images_batch)):
    risk, dist = get_risk_and_distance(predictions[i])
    plt.imshow(images_batch[i])
    plt.title(f"Predicted: {risk}, Safe Distance: {dist}m")
    plt.axis('off')
    plt.show()


print("Done.")
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report
import random
import os

# Set random seeds for reproducibility
tf.random.set_seed(42)
np.random.seed(42)
random.seed(42)

# Parameters
IMG_SIZE = 64
BATCH_SIZE = 16
EPOCHS = 10
NUM_CLASSES = 2  # Low risk, High risk

LOW_RISK_CLASSES = ['Forest', 'Pasture', 'PermanentCrop', 'River', 'SeaLake']
HIGH_RISK_CLASSES = ['AnnualCrop', 'HerbaceousVegetation', 'Highway', 'Industrial', 'Residential']

SAFE_DISTANCE = {
    'Low Risk': 50,
    'High Risk': 200
}

print("Loading EuroSAT dataset from TensorFlow Datasets...")
import tensorflow_datasets as tfds

# Load EuroSAT dataset, RGB images
dataset, info = tfds.load('eurosat/rgb', split='train', with_info=True, as_supervised=True)
class_names = info.features['label'].names
class_name_to_label = {name: idx for idx, name in enumerate(class_names)}
low_risk_labels = [class_name_to_label[c] for c in LOW_RISK_CLASSES if c in class_name_to_label]
high_risk_labels = [class_name_to_label[c] for c in HIGH_RISK_CLASSES if c in class_name_to_label]

low_risk_ds = dataset.filter(lambda img, label: tf.reduce_any(label == low_risk_labels))
high_risk_ds = dataset.filter(lambda img, label: tf.reduce_any(label == high_risk_labels))

def sample_n(ds, n):
    ds = ds.shuffle(1000, seed=42)
    return ds.take(n)

low_risk_sampled = sample_n(low_risk_ds, 100)
high_risk_sampled = sample_n(high_risk_ds, 100)
full_ds = low_risk_sampled.concatenate(high_risk_sampled)

# Data augmentation
data_augmentation = keras.Sequential([
    layers.Rescaling(1./255),
    layers.RandomFlip("horizontal_and_vertical"),
    layers.RandomRotation(0.2),
    layers.RandomZoom(0.2),
])

def preprocess_img_label(image, label):
    image = tf.image.resize(image, [IMG_SIZE, IMG_SIZE])
    image = data_augmentation(image)
    label_binary = tf.where(tf.reduce_any(tf.equal(label, low_risk_labels)), 0, 1)
    return image, label_binary

full_ds = full_ds.map(preprocess_img_label, num_parallel_calls=tf.data.AUTOTUNE)
full_ds = full_ds.shuffle(200, seed=42)

train_count = 170
val_count = 30
train_ds = full_ds.take(train_count).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)
val_ds = full_ds.skip(train_count).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)

print(f"Using 200 images: {train_count} train, {val_count} val")

class PatchExtract(layers.Layer):
    def __init__(self, patch_size):
        super().__init__()
        self.patch_size = patch_size

    def call(self, images):
        batch_size = tf.shape(images)[0]
        patches = tf.image.extract_patches(
            images=images,
            sizes=[1, self.patch_size, self.patch_size, 1],
            strides=[1, self.patch_size, self.patch_size, 1],
            rates=[1, 1, 1, 1],
            padding='VALID'
        )
        patch_dims = patches.shape[-1]
        patches = tf.reshape(patches, [batch_size, -1, patch_dims])
        return patches

class PatchEncoder(layers.Layer):
    def __init__(self, num_patches, projection_dim):
        super().__init__()
        self.projection = layers.Dense(units=projection_dim)
        self.position_embedding = layers.Embedding(input_dim=num_patches, output_dim=projection_dim)

    def call(self, patch):
        positions = tf.range(start=0, limit=self.position_embedding.input_dim, delta=1)
        encoded = self.projection(patch) + self.position_embedding(positions)
        return encoded

def create_vit_classifier(input_shape=(IMG_SIZE, IMG_SIZE, 3), patch_size=8, projection_dim=64, num_heads=2, transformer_layers=4, mlp_head_units=[128, 64], num_classes=NUM_CLASSES):
    inputs = layers.Input(shape=input_shape)
    patches = PatchExtract(patch_size)(inputs)
    num_patches = (input_shape[0] // patch_size) * (input_shape[1] // patch_size)
    encoded_patches = PatchEncoder(num_patches, projection_dim)(patches)

    for _ in range(transformer_layers):
        x1 = layers.LayerNormalization(epsilon=1e-6)(encoded_patches)
        attention_output = layers.MultiHeadAttention(num_heads=num_heads, key_dim=projection_dim, dropout=0.1)(x1, x1)
        x2 = layers.Add()([attention_output, encoded_patches])
        x3 = layers.LayerNormalization(epsilon=1e-6)(x2)
        mlp_output = keras.Sequential([
            layers.Dense(units=projection_dim * 2, activation=tf.nn.gelu),
            layers.Dense(units=projection_dim, activation=tf.nn.gelu),
        ])(x3)
        encoded_patches = layers.Add()([mlp_output, x2])

    representation = layers.LayerNormalization(epsilon=1e-6)(encoded_patches)
    representation = layers.Flatten()(representation)
    representation = layers.Dropout(0.5)(representation)

    for units in mlp_head_units:
        representation = layers.Dense(units, activation=tf.nn.gelu)(representation)
        representation = layers.Dropout(0.3)(representation)

    logits = layers.Dense(num_classes, activation='softmax')(representation)
    return keras.Model(inputs=inputs, outputs=logits)

lr_schedule = keras.optimizers.schedules.ExponentialDecay(initial_learning_rate=1e-3, decay_steps=100, decay_rate=0.96, staircase=True)
model = create_vit_classifier()
model.compile(optimizer=keras.optimizers.Adam(learning_rate=lr_schedule), loss='sparse_categorical_crossentropy', metrics=['accuracy'])

callbacks = [
    keras.callbacks.EarlyStopping(patience=3, restore_best_weights=True),
    keras.callbacks.ModelCheckpoint("best_model.h5", save_best_only=True)
]

print(model.summary())

print("Starting model training...")
history = model.fit(train_ds, epochs=EPOCHS, validation_data=val_ds, callbacks=callbacks)

def plot_training(history):
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Train')
    plt.plot(history.history['val_accuracy'], label='Val')
    plt.title('Accuracy')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Train')
    plt.plot(history.history['val_loss'], label='Val')
    plt.title('Loss')
    plt.legend()
    plt.show()

plot_training(history)

# Evaluate
print("\nValidation Set Evaluation:")
y_true, y_pred = [], []
for images, labels in val_ds:
    preds = model.predict(images)
    y_true.extend(labels.numpy())
    y_pred.extend(np.argmax(preds, axis=1))

cm = confusion_matrix(y_true, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Low', 'High'], yticklabels=['Low', 'High'])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()

print(classification_report(y_true, y_pred, target_names=['Low Risk', 'High Risk']))

# Sample prediction display
def get_risk_and_distance(pred_probs):
    pred_label = np.argmax(pred_probs)
    risk = "Low Risk" if pred_label == 0 else "High Risk"
    distance = SAFE_DISTANCE[risk]
    return risk, distance

print("\nSample predictions on validation set:")
val_iter = iter(val_ds.unbatch().batch(5))
images_batch, labels_batch = next(val_iter)
predictions = model.predict(images_batch)
for i in range(len(images_batch)):
    risk, dist = get_risk_and_distance(predictions[i])
    plt.imshow(images_batch[i])
    plt.title(f"Predicted: {risk}, Safe Distance: {dist}m")
    plt.axis('off')
    plt.show()

print("Done.")
```

```python
import cv2

# Select the transformer block and layer for Grad-CAM
# We'll use the last attention output in the ViT
target_layer_name = None
for layer in reversed(model.layers):
    if isinstance(layer, tf.keras.layers.MultiHeadAttention):
        target_layer_name = layer.name
        break

if target_layer_name is None:
    raise ValueError("No MultiHeadAttention layer found for Grad-CAM.")

# Grad-CAM function
def make_gradcam_heatmap(img_array, model, last_attention_layer_name):
    grad_model = tf.keras.models.Model(
        [model.inputs],
        [model.get_layer(last_attention_layer_name).output, model.output]
    )

    with tf.GradientTape() as tape:
        attention_output, preds = grad_model(img_array)
        class_idx = tf.argmax(preds[0])
        loss = preds[:, class_idx]

    grads = tape.gradient(loss, attention_output)

    # Global average pooling
    pooled_grads = tf.reduce_mean(grads, axis=1)

    # Weight the attention output
    attention_output = attention_output[0]
    pooled_grads = pooled_grads[0]
    cam = tf.reduce_sum(tf.multiply(pooled_grads, attention_output), axis=-1)

    # Reshape CAM to image shape
    heatmap = tf.reshape(cam, [IMG_SIZE // 8, IMG_SIZE // 8])
    heatmap = tf.maximum(heatmap, 0) / tf.math.reduce_max(heatmap)
    heatmap = tf.image.resize(tf.expand_dims(heatmap, -1), [IMG_SIZE, IMG_SIZE])
    heatmap = tf.squeeze(heatmap)
    return heatmap.numpy()

# Overlay heatmap
def superimpose_heatmap(img, heatmap, alpha=0.4):
    heatmap = np.uint8(255 * heatmap)
    heatmap_colored = cv2.applyColorMap(heatmap, cv2.COLORMAP_JET)
    img = np.uint8(img * 255)
    superimposed_img = cv2.addWeighted(img, 1 - alpha, heatmap_colored, alpha, 0)
    return superimposed_img

# Run Grad-CAM on a few samples
print("\nGrad-CAM Visualizations:")
val_iter = iter(val_ds.unbatch().batch(5))
images_batch, labels_batch = next(val_iter)
predictions = model.predict(images_batch)

for i in range(len(images_batch)):
    img = images_batch[i:i+1]
    heatmap = make_gradcam_heatmap(img, model, target_layer_name)
    superimposed_img = superimpose_heatmap(img[0].numpy(), heatmap)

    risk, dist = get_risk_and_distance(predictions[i])

    plt.figure(figsize=(8, 4))
    plt.subplot(1, 2, 1)
    plt.imshow(images_batch[i])
    plt.title(f"Input Image\n{risk}, Safe Distance: {dist}m")
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(superimposed_img)
    plt.title("Grad-CAM")
    plt.axis('off')
    plt.tight_layout()
    plt.show()
```

```python
# Vision Transformer-based Disaster Risk Classifier with Enhancements

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report
import random
import os

# Reproducibility
tf.random.set_seed(42)
np.random.seed(42)
random.seed(42)

# Constants
IMG_SIZE = 64
BATCH_SIZE = 16
EPOCHS = 15
NUM_CLASSES = 1  # Binary classification with sigmoid

LOW_RISK_CLASSES = ['Forest', 'Pasture', 'PermanentCrop', 'River', 'SeaLake']
HIGH_RISK_CLASSES = ['AnnualCrop', 'HerbaceousVegetation', 'Highway', 'Industrial', 'Residential']

SAFE_DISTANCE = {
    'Low Risk': 50,
    'High Risk': 200
}

print("Loading EuroSAT dataset...")
import tensorflow_datasets as tfds
dataset, info = tfds.load('eurosat/rgb', split='train', with_info=True, as_supervised=True)
class_names = info.features['label'].names
class_name_to_label = {name: idx for idx, name in enumerate(class_names)}
low_risk_labels = [class_name_to_label[c] for c in LOW_RISK_CLASSES if c in class_name_to_label]
high_risk_labels = [class_name_to_label[c] for c in HIGH_RISK_CLASSES if c in class_name_to_label]

def is_low_risk(label):
    return tf.reduce_any(tf.equal(label, low_risk_labels))

def is_high_risk(label):
    return tf.reduce_any(tf.equal(label, high_risk_labels))

low_risk_ds = dataset.filter(lambda img, label: is_low_risk(label))
high_risk_ds = dataset.filter(lambda img, label: is_high_risk(label))

def sample_n(ds, n):
    return ds.shuffle(1000, seed=42).take(n)

low_risk_sampled = sample_n(low_risk_ds, 1000)
high_risk_sampled = sample_n(high_risk_ds, 1000)
full_ds = low_risk_sampled.concatenate(high_risk_sampled)

# Data Augmentation
data_augmentation = keras.Sequential([
    layers.Rescaling(1./255),
    layers.RandomFlip("horizontal_and_vertical"),
    layers.RandomRotation(0.2),
    layers.RandomZoom(0.2),
])

def preprocess_img_label(image, label):
    image = tf.image.resize(image, [IMG_SIZE, IMG_SIZE])
    image = data_augmentation(image)
    label_binary = tf.where(is_low_risk(label), 0.0, 1.0)
    return image, label_binary

full_ds = full_ds.map(preprocess_img_label, num_parallel_calls=tf.data.AUTOTUNE)
full_ds = full_ds.shuffle(2000, seed=42)

# Split Dataset
train_size = 1400
val_size = 300
train_ds = full_ds.take(train_size).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)
val_ds = full_ds.skip(train_size).take(val_size).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)
test_ds = full_ds.skip(train_size + val_size).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)

# ViT Components
class PatchExtract(layers.Layer):
    def __init__(self, patch_size):
        super().__init__()
        self.patch_size = patch_size

    def call(self, images):
        batch_size = tf.shape(images)[0]
        patches = tf.image.extract_patches(
            images=images,
            sizes=[1, self.patch_size, self.patch_size, 1],
            strides=[1, self.patch_size, self.patch_size, 1],
            rates=[1, 1, 1, 1],
            padding='VALID'
        )
        patch_dims = patches.shape[-1]
        return tf.reshape(patches, [batch_size, -1, patch_dims])

class PatchEncoder(layers.Layer):
    def __init__(self, num_patches, projection_dim):
        super().__init__()
        self.projection = layers.Dense(projection_dim)
        self.position_embedding = layers.Embedding(input_dim=num_patches, output_dim=projection_dim)

    def call(self, patch):
        positions = tf.range(start=0, limit=self.position_embedding.input_dim, delta=1)
        return self.projection(patch) + self.position_embedding(positions)

def create_vit_classifier(input_shape=(IMG_SIZE, IMG_SIZE, 3), patch_size=8, projection_dim=64, num_heads=4, transformer_layers=4, mlp_head_units=[128, 64]):
    inputs = layers.Input(shape=input_shape)
    patches = PatchExtract(patch_size)(inputs)
    num_patches = (input_shape[0] // patch_size) ** 2
    encoded_patches = PatchEncoder(num_patches, projection_dim)(patches)

    for _ in range(transformer_layers):
        x1 = layers.LayerNormalization(epsilon=1e-6)(encoded_patches)
        attention_output = layers.MultiHeadAttention(num_heads=num_heads, key_dim=projection_dim, dropout=0.1)(x1, x1)
        x2 = layers.Add()([attention_output, encoded_patches])
        x3 = layers.LayerNormalization(epsilon=1e-6)(x2)
        mlp_output = keras.Sequential([
            layers.Dense(units=projection_dim * 2, activation=tf.nn.gelu),
            layers.Dropout(0.1),
            layers.Dense(units=projection_dim),
        ])(x3)
        encoded_patches = layers.Add()([mlp_output, x2])

    representation = layers.LayerNormalization(epsilon=1e-6)(encoded_patches)
    representation = layers.Flatten()(representation)
    representation = layers.Dropout(0.5)(representation)
    for units in mlp_head_units:
        representation = layers.Dense(units, activation=tf.nn.gelu)(representation)
        representation = layers.Dropout(0.3)(representation)

    logits = layers.Dense(1, activation='sigmoid')(representation)
    return keras.Model(inputs=inputs, outputs=logits)

# Compile Model
model = create_vit_classifier()
lr_schedule = keras.optimizers.schedules.ExponentialDecay(1e-3, decay_steps=200, decay_rate=0.9, staircase=True)
model.compile(optimizer=keras.optimizers.Adam(lr_schedule),
              loss='binary_crossentropy',
              metrics=['accuracy', keras.metrics.AUC(), keras.metrics.Precision(), keras.metrics.Recall()])

callbacks = [
    keras.callbacks.EarlyStopping(patience=5, restore_best_weights=True),
    keras.callbacks.ModelCheckpoint("best_vit_model.h5", save_best_only=True),
    keras.callbacks.TensorBoard(log_dir="./logs")
]

# Train Model
history = model.fit(train_ds, validation_data=val_ds, epochs=EPOCHS, callbacks=callbacks)

# Plot Training History
def plot_training(history):
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Train')
    plt.plot(history.history['val_accuracy'], label='Val')
    plt.title('Accuracy')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Train')
    plt.plot(history.history['val_loss'], label='Val')
    plt.title('Loss')
    plt.legend()
    plt.show()

plot_training(history)

# Evaluate on Test Set
y_true, y_pred = [], []
for images, labels in test_ds:
    preds = model.predict(images)
    y_true.extend(labels.numpy())
    y_pred.extend((preds > 0.5).astype(int))

cm = confusion_matrix(y_true, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Low', 'High'], yticklabels=['Low', 'High'])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()

print(classification_report(y_true, y_pred, target_names=['Low Risk', 'High Risk']))

# Grad-CAM Visualization
def make_gradcam_heatmap(img_array, model, last_conv_layer_name):
    grad_model = tf.keras.models.Model(
        [model.inputs], [model.get_layer(last_conv_layer_name).output, model.output]
    )
    with tf.GradientTape() as tape:
        conv_outputs, predictions = grad_model(img_array)
        loss = predictions[:, 0]
    grads = tape.gradient(loss, conv_outputs)[0]
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1))
    conv_outputs = conv_outputs[0]
    heatmap = tf.reduce_sum(tf.multiply(pooled_grads, conv_outputs), axis=-1)
    heatmap = tf.maximum(heatmap, 0) / tf.math.reduce_max(heatmap)
    return heatmap.numpy()

print("\nGrad-CAM predictions:")
sample_batch = next(iter(test_ds.unbatch().batch(1)))
image, label = sample_batch
prediction = model.predict(image)
heatmap = make_gradcam_heatmap(image, model, last_conv_layer_name='dropout')  # update this if using a conv stem

import cv2
img = image[0].numpy()
img_rescaled = (img * 255).astype(np.uint8)
heatmap_resized = cv2.resize(heatmap, (img.shape[1], img.shape[0]))
heatmap_colored = cv2.applyColorMap(np.uint8(255 * heatmap_resized), cv2.COLORMAP_JET)
superimposed_img = cv2.addWeighted(img_rescaled, 0.6, heatmap_colored, 0.4, 0)

plt.imshow(superimposed_img)
risk_label = "Low Risk" if prediction < 0.5 else "High Risk"
plt.title(f"Prediction: {risk_label}, Distance: {SAFE_DISTANCE[risk_label]}m")
plt.axis('off')
plt.show()

print("Done.")
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
import matplotlib.pyplot as plt
import random
import os

# Set random seeds for reproducibility
tf.random.set_seed(42)
np.random.seed(42)
random.seed(42)

# Parameters
IMG_SIZE = 64
BATCH_SIZE = 16
EPOCHS = 20  # Increased epochs for better training
NUM_CLASSES = 2  # Low risk, High risk

# Map some EuroSAT classes to disaster risk level - for illustration
LOW_RISK_CLASSES = ['Forest', 'Pasture', 'PermanentCrop', 'River', 'SeaLake']
HIGH_RISK_CLASSES = ['AnnualCrop', 'HerbaceousVegetation', 'Highway', 'Industrial', 'Residential']

# For safe distances in meters for each risk level:
SAFE_DISTANCE = {
    'Low Risk': 50,   # safe distance in meters
    'High Risk': 200
}

print("Loading EuroSAT dataset from TensorFlow Datasets...")

import tensorflow_datasets as tfds

# Load EuroSAT dataset, RGB images
dataset, info = tfds.load('eurosat/rgb', split='train', with_info=True, as_supervised=True)

# Reverse mapping from class name to label
class_names = info.features['label'].names
class_name_to_label = {name: idx for idx, name in enumerate(class_names)}

# Filter classes
low_risk_labels = [class_name_to_label[c] for c in LOW_RISK_CLASSES if c in class_name_to_label]
high_risk_labels = [class_name_to_label[c] for c in HIGH_RISK_CLASSES if c in class_name_to_label]

# Separate dataset into low and high risk class subsets
low_risk_ds = dataset.filter(lambda img, label: tf.reduce_any(label == low_risk_labels))
high_risk_ds = dataset.filter(lambda img, label: tf.reduce_any(label == high_risk_labels))

# Function to sample N examples from dataset
def sample_n(ds, n):
    ds = ds.shuffle(1000, seed=42)
    return ds.take(n)

# Take ~200 images from low risk and ~200 from high risk
low_risk_count = 200
high_risk_count = 200
low_risk_sampled = sample_n(low_risk_ds, low_risk_count)
high_risk_sampled = sample_n(high_risk_ds, high_risk_count)

# Merge datasets
full_ds = low_risk_sampled.concatenate(high_risk_sampled)

# Function to preprocess and label disaster risk
def preprocess_img_label(image, label):
    # Resize images to IMG_SIZE x IMG_SIZE
    image = tf.image.resize(image, [IMG_SIZE, IMG_SIZE])
    image = tf.cast(image, tf.float32) / 255.0  # normalize to [0,1]

    # Map the original label to 0 (low risk) or 1 (high risk)
    label_binary = tf.py_function(lambda x: 0 if x in low_risk_labels else 1, [label], tf.int64)
    label_binary.set_shape([])

    return image, label_binary

# Map preprocessing
full_ds = full_ds.map(preprocess_img_label, num_parallel_calls=tf.data.AUTOTUNE)
full_ds = full_ds.shuffle(200, seed=42)

# Data Augmentation
data_augmentation = keras.Sequential([
    layers.RandomFlip("horizontal_and_vertical"),
    layers.RandomRotation(0.2),
    layers.RandomZoom(0.1),
])

# Split into train and validation (85% / 15%)
total_count = low_risk_count + high_risk_count
train_count = int(total_count * 0.85)
val_count = total_count - train_count

train_ds = full_ds.take(train_count).map(lambda x, y: (data_augmentation(x, training=True), y)).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)
val_ds = full_ds.skip(train_count).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)

print(f"Using {total_count} images: {train_count} train, {val_count} val")

# Define Vision Transformer components (same as before)
class PatchExtract(layers.Layer):
    def __init__(self, patch_size):
        super().__init__()
        self.patch_size = patch_size

    def call(self, images):
        batch_size = tf.shape(images)[0]
        patches = tf.image.extract_patches(
            images=images,
            sizes=[1, self.patch_size, self.patch_size, 1],
            strides=[1, self.patch_size, self.patch_size, 1],
            rates=[1,1,1,1],
            padding='VALID'
        )
        patch_dims = patches.shape[-1]
        patches = tf.reshape(patches, [batch_size, -1, patch_dims])
        return patches

class PatchEncoder(layers.Layer):
    def __init__(self, num_patches, projection_dim):
        super().__init__()
        self.num_patches = num_patches
        self.projection = layers.Dense(units=projection_dim)
        self.position_embedding = layers.Embedding(
            input_dim=num_patches, output_dim=projection_dim
        )

    def call(self, patch):
        positions = tf.range(start=0, limit=self.num_patches, delta=1)
        encoded = self.projection(patch) + self.position_embedding(positions)
        return encoded

def create_vit_classifier(
    input_shape=(IMG_SIZE, IMG_SIZE, 3),
    patch_size=8,
    projection_dim=64,
    num_heads=4,  # Increased number of heads
    transformer_layers=6,  # Increased number of transformer layers
    mlp_head_units=[128, 64],
    num_classes=NUM_CLASSES,
):
    inputs = layers.Input(shape=input_shape)
    patches = PatchExtract(patch_size)(inputs)
    num_patches = (input_shape[0] // patch_size) * (input_shape[1] // patch_size)
    encoded_patches = PatchEncoder(num_patches, projection_dim)(patches)

    for _ in range(transformer_layers):
        x1 = layers.LayerNormalization(epsilon=1e-6)(encoded_patches)
        attention_output = layers.MultiHeadAttention(
            num_heads=num_heads, key_dim=projection_dim, dropout=0.1
        )(x1, x1)
        x2 = layers.Add()([attention_output, encoded_patches])
        x3 = layers.LayerNormalization(epsilon=1e-6)(x2)
        mlp_output = keras.Sequential(
            [
                layers.Dense(units=projection_dim * 2, activation=tf.nn.gelu),
                layers.Dense(units=projection_dim, activation=tf.nn.gelu),
            ]
        )(x3)
        encoded_patches = layers.Add()([mlp_output, x2])

    representation = layers.LayerNormalization(epsilon=1e-6)(encoded_patches)
    representation = layers.Flatten()(representation)
    representation = layers.Dropout(0.5)(representation)

    for units in mlp_head_units:
        representation = layers.Dense(units, activation=tf.nn.gelu)(representation)
        representation = layers.Dropout(0.3)(representation)

    logits = layers.Dense(num_classes, activation='softmax')(representation)

    model = keras.Model(inputs=inputs, outputs=logits)
    return model

# Create the model
model = create_vit_classifier()
model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=1e-4),  # Adjusted learning rate
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy'],
)

print(model.summary())

# Early stopping callback
early_stopping = keras.callbacks.EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

# Train the model
print("Starting model training...")
history = model.fit(train_ds, epochs=EPOCHS, validation_data=val_ds, callbacks=[early_stopping])

# Function to map predicted label to risk level and safe distance
def get_risk_and_distance(pred_probs):
    pred_label = np.argmax(pred_probs)
    if pred_label == 0:
        risk = "Low Risk"
    else:
        risk = "High Risk"
    distance = SAFE_DISTANCE[risk]
    return risk, distance

# Evaluate some samples from validation set and display
print("\nSample predictions on validation set:")
val_iter = iter(val_ds.unbatch().batch(5))
images_batch, labels_batch = next(val_iter)
predictions = model.predict(images_batch)
for i in range(len(images_batch)):
    risk, dist = get_risk_and_distance(predictions[i])
    plt.imshow(images_batch[i])
    plt.title(f"Predicted: {risk}, Safe Distance: {dist}m")
    plt.axis('off')
    plt.show()

print("Done.")
```
